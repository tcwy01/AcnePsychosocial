```python
#Replacing height values that are higher than 251cm (recorded height of tallest living people) and lower than 130cm (just less than average height of 10 year old female) as NaN 
# - (https://www.disabled-world.com/calculators-charts/height-weight-teens.php#fc)
import numpy as np
acne_psychosocial['Height'] = acne_psychosocial['Height'].apply(
    lambda x:x if x < 251 else np.nan)
acne_psychosocial['Height'] = acne_psychosocial['Height'].apply(
    lambda x:x if x > 130 else np.nan)

#Replacing weight values that are higher than 650kg (recorded weight of heaviest living person) and lower than 25kg (just less than average weight of 10 year old female) as NaN
# - (https://www.disabled-world.com/calculators-charts/height-weight-teens.php#fc)
acne_psychosocial['Weight'] = acne_psychosocial['Weight'].apply(
    lambda x:x if x < 650 else np.nan)
acne_psychosocial['Weight'] = acne_psychosocial['Weight'].apply(
    lambda x:x if x > 25 else np.nan)
acne_psychosocial.describe()
```

```python
#Converting Weight and Height cols to BMI, as BMINew had substantial missingness
acne_psychosocial['BMI'] = acne_psychosocial['Weight'] / acne_psychosocial['Height'].div(100).pow(2)
acne_psychosocial['BMI'] = acne_psychosocial['BMI'].round(1)
#acne_psychosocial.drop(['Height', 'Weight'], axis=1, inplace=True)
acne_psychosocial
```

```python
#Loading in numpy
import numpy as np

#Adding column for adolescent (0) or adult (1), based on age:
# - According to the World Health Organisation, the term 'young people' are individuals between 10 to 24 years of age
# - and adult female acne is defined as impacting women over the age of 25 
# - (Bagatin E, Freitas THP, Rivitti-Machado MC, Machado MCR, Ribeiro BM, Nunes S, Rocha MADD. Adult female acne: a guide to clinical practice. An Bras Dermatol. 2019 Jan-Feb;94(1):62-75. doi: 10.1590/abd1806-4841.20198203. Erratum in: An Bras Dermatol. 2019 Mar-Apr;94(2):255. doi: 10.1590/abd1806-4841.2019940202. Machado MCR [corrected to Rivitti-Machado MC]. PMID: 30726466; PMCID: PMC6360964.)

#Defining function to determine if adolescent or adult
def yes_adult(age):
    if pd.isna(age): 
        return np.nan
    elif age > 25: 
        return 1
    else:
        return 0

#Applying function 
acne_psychosocial['Yes_adult'] = acne_psychosocial['Age'].apply(yes_adult)
acne_psychosocial
```

```python
#Checking if binary function for adult or adolescent was implemented
acne_psychosocial.loc[:, ['Age', 'Yes_adult']]
```

```python
#Adding column for low (0) and middle-to-upper (1) socioeconomic income, based on monthly income:
# - According to the Office for National Statistics, median household income was estimated at £32,300 in 2022 
# - (https://www.ons.gov.uk/peoplepopulationandcommunity/personalandhouseholdfinances/incomeandwealth/bulletins/householddisposableincomeandinequality/financialyearending2022)
# - Low income is defined as households earning below 60% of the median, equating to £1,615 per month 
# - Hence, 'Monthly_income_1000to1500' will be considered low income, whilst 'Monthly_income_1500to2000' will be considered middle-to-upper income

#Defining incomes columns 
incomes = ['Monthly_Income_less_than_10',
           'Monthly_Income_10to50', 
           'Monthly_Income_50to100',
           'Monthly_Income_100to250', 
           'Monthly_Income_250to500',
           'Monthly_Income_500to1000', 
           'Monthly_Income_1000to1500',
           'Monthly_Income_1500to2000', 
           'Monthly_Income_2000to2500', 
           'Monthly_Income_2500to3000',
           'Monthly_Income_3000to3500', 
           'Monthly_Income_more_than3500']

#Defining low income and middle-to-upper income
low_income = ['Monthly_Income_less_than_10', 
              'Monthly_Income_10to50', 
              'Monthly_Income_50to100', 
              'Monthly_Income_100to250', 
              'Monthly_Income_250to500', 
              'Monthly_Income_500to1000', 
              'Monthly_Income_1000to1500']
middle_upper_income = ['Monthly_Income_1500to2000', 
                       'Monthly_Income_2000to2500', 
                       'Monthly_Income_2500to3000', 
                       'Monthly_Income_3000to3500', 
                       'Monthly_Income_more_than3500']

#Defining function to determine if low or middle-to-upper
def socioeconomic_status(income):
    if income[low_income].sum() > 0: 
        return 0
    elif income[middle_upper_income].sum() > 0: 
        return 1
    elif pd.isna(income[incomes]).all():
        return np.nan
    else: 
        return np.nan

#Applying function 
acne_psychosocial['Socioeconomic_status'] = acne_psychosocial.apply(socioeconomic_status, axis=1)

#Dropping monthly income columns 
acne_psychosocial.drop(incomes, axis=1, inplace=True)
acne_psychosocial
```

```python
#Loading in missingno
#! pip install missingno
import missingno as msno

#Inspecting missing values 
msno.matrix(acne_psychosocial)
#Calculating percentage missingness 
missingness = acne_psychosocial.isnull().mean() * 100

#Printing columns with missingness greater than 90%
severe_missing = missingness[missingness > 90].index
severe_missing

#Dropping columns with missingness greater than 90%
# - Substantial missingness may reduce statistical power if used in the analysis, 
# - and cause large inaccuracies if imputation were to be carried out
acne_psychosocial.drop(severe_missing, axis=1, inplace=True)
acne_psychosocial
```

```python
#Counting number of adolescents and adults
Adolescents_count = (acne_psychosocial['Age'] < 25).sum()
Adults_count = (acne_psychosocial['Age'] >= 25).sum()
print("The number of adolescent patients: " + str(Adolescents_count))
print("The number of adult patients: " + str(Adults_count))
```

```python
#Classifying adult acne based on age, onset and duration cols 
# - late-onset adult acne (1): occurring only after the age of 25
# - persistent adult acne (2): persisting continously from adolescence 
# - recurrent adult acne (3): present in adolescence, stops for a variable time period, and returns in adulthood
#(Bagatin E, Adult female acne, doi: 10.1590/abd1806-4841.2019940202)

#Defining function to classify adult acne 
def adult_acne_type(row):
    onset_age = row['ONSET']
    current_age = row['Age']
    duration = row['Duration']

    if onset_age > 25:
        return 1
    elif onset_age <= 25 and current_age > 25 and (current_age - onset_age) == duration:
        return 2
    elif onset_age <= 25 and current_age > 25 and (current_age - onset_age) > duration:
        return 3
    else: return np.nan

#Applying function
acne_psychosocial['Type_adultAcne'] = acne_psychosocial.apply(adult_acne_type, axis=1)
acne_psychosocial
```

```python
#Counting numbers of each category of adult acne 
late_onset_count = (acne_psychosocial['Type_adultAcne']==1.0).sum()
persistent_count = (acne_psychosocial['Type_adultAcne']==2.0).sum()
recurrent_count = (acne_psychosocial['Type_adultAcne']==3.0).sum()

#Printing values
print("Number of patients with late-onset adult acne: " + str(late_onset_count))
print("Number of patients with persistent adult acne: " + str(persistent_count))
print("Number of patients with recurrent adult acne: " + str(recurrent_count))
```

```python
#Counting PHQ-9 scores based on category (from minimal to severe)
#(https://www.hiv.uw.edu/page/mental-health-screening/phq-9)
phq9_minimal = ((acne_psychosocial['ResultsPHQ9']>=0)&(acne_psychosocial['ResultsPHQ9']<=4)).sum()
phq9_mild = ((acne_psychosocial['ResultsPHQ9']>=5)&(acne_psychosocial['ResultsPHQ9']<=9)).sum()
phq9_moderate = ((acne_psychosocial['ResultsPHQ9']>=10)&(acne_psychosocial['ResultsPHQ9']<=14)).sum()
phq9_modSevere = ((acne_psychosocial['ResultsPHQ9']>=15)&(acne_psychosocial['ResultsPHQ9']<=19)).sum()
phq9_severe = ((acne_psychosocial['ResultsPHQ9']>=20)&(acne_psychosocial['ResultsPHQ9']<=27)).sum()

#Printing values
print("Number of patients with none/minimal PHQ9 score: " + str(phq9_minimal))
print("Number of patients with mild PHQ9 score: " + str(phq9_mild))
print("Number of patients with moderate PHQ9 score: " + str(phq9_moderate))
print("Number of patients with moderately severe PHQ9 score: " + str(phq9_modSevere))
print("Number of patients with severe PHQ9 score: " + str(phq9_severe))
```
